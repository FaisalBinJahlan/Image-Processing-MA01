```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure, img_as_float
```

```python
img = data.moon()
plt.imshow(img, cmap='gray')
plt.title("Original Moon Image")
plt.axis('off')
plt.show()
```

```python
#task 1

p3, p80 = np.percentile(img, (3, 80))
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

plt.imshow(img_rescale, cmap='gray')
plt.title("Contrast Stretch (3% - 80%)")
plt.axis('off')
plt.show()
```

```python
#Task 1 plot
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(img.flat, bins=256, range=(0,255))
plt.title("Original Histogram")

plt.subplot(1,2,2)
plt.hist(img_rescale.flat, bins=256, range=(0,255))
plt.title("Rescaled Histogram (3%-80%)")

plt.show()
```

```python
#Task2
img = data.moon()
img_equalized = exposure.equalize_hist(img)

plt.imshow(img_equalized, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')
plt.show()
```

```python
#Task 2 plot
plt.hist(img_equalized.flat, bins=256)
plt.title("Equalized Histogram")
plt.show()
```

```python
#Task 3
source = data.chelsea()
reference = data.rocket()

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(source)
plt.title("Source - Chelsea")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(reference)
plt.title("Reference - Rocket")
plt.axis('off')

plt.show()
```

```python
#Task3 plot
matched = exposure.match_histograms(source, reference, channel_axis=-1)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(source)
plt.title("Original Source")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(matched)
plt.title("After Histogram Matching")
plt.axis('off')

plt.show()
```
