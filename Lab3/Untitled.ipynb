```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_color = cv2.imread('images/image1.jpeg')
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

img_gray = cv2.cvtColor(img_color, cv2.COLOR_RGB2GRAY)
```

```python
#Task 1.1
height, width = img_color.shape[:2]
resized_img = cv2.resize(img_color, (width*2, height*2))

plt.imshow(resized_img)
plt.title("Increased Size")
plt.axis('off')
plt.show()
```

```python
#Task 1.2
(h, w) = img_color.shape[:2]
center = (w//2, h//2)

rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated_img = cv2.warpAffine(img_color, rotation_matrix, (w, h))

plt.imshow(rotated_img)
plt.title("Rotated 120 Degrees")
plt.axis('off')
plt.show()
```

```python
#Task 1.3
rows, cols, ch = img_color.shape

shear_matrix = np.float32([[1, 0.5, 0],
                           [0, 1, 0]])

sheared_img = cv2.warpAffine(img_color, shear_matrix, (cols, rows))

plt.imshow(sheared_img)
plt.title("Sheared Image")
plt.axis('off')
plt.show()
```

```python
#Task 2.1
negative = 255 - img_gray

plt.imshow(negative, cmap='gray')
plt.title("Negative")
plt.axis('off')
plt.show()
```

```python
img_float = img_gray.astype(np.float32)

c = 255 / np.log(1 + np.max(img_float))
log_img = c * np.log(1 + img_float)

log_img = np.clip(log_img, 0, 255)
log_img = log_img.astype(np.uint8)

plt.imshow(log_img, cmap='gray')
plt.title("Log Transformation")
plt.axis('off')
plt.show()
```

```python
#Task 2.3
gamma = 0.5
power_img = np.array(255 * (img_gray / 255) ** gamma, dtype='uint8')

plt.imshow(power_img, cmap='gray')
plt.title("Power Law")
plt.axis('off')
plt.show()
```
